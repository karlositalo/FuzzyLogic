```python
mat = np.array([[1.,0.6,0.4,0.3],[0.7,1.,0.1,0.4],[0.2,0.4,1.,0.3],[0.5,0.3,0.4,1.]])
#Matriz Condicional
def MaCond(A):
    n = A.shape[0]
    C = np.zeros((n,n))
    for i in range(n):
        for j in range(n):
            C[i,j] = A[j,i] / max(A[j,i],A[i,j])
    return C

#Vetor de ordens
def V_Ordem(M):
    n = M.shape[0]
    v = np.zeros((n))
    for i in range(n):
        v[i] = min(M[i,:])
    return v
```

```python
#forma 01
matCond = MaCond(mat)
resultados = V_Ordem(matCond)
resultados
```

```python
def ProMarg(Mat,P): 
    l,c = Mat.shape
    Pm = np.zeros((l))
    for j in range(l):
        for i in range(l):
            Pm[j] = Pm[j] + Mat[i,j]*P[i]
    return Pm
```
